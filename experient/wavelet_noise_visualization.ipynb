# üåÄ Wavelet Noise Visualization

This notebook visualizes the outputs from the **Wavelet Noise** implementation based on Cook & DeRose (2005).

You can compare Wavelet vs. Perlin noise patterns, analyze frequency bands, and view spectrum images.

```python
# Import required libraries
import matplotlib.pyplot as plt
import os
from PIL import Image

# Set image folder path
img_dir = './'  # Modify this if images are in a subfolder

# Helper to load image
def load_image(name):
    path = os.path.join(img_dir, name)
    return Image.open(path) if os.path.exists(path) else None

def show_side_by_side(img1, img2, title1, title2):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    for ax, img, title in zip(axes, [img1, img2], [title1, title2]):
        ax.imshow(img, cmap='gray')
        ax.set_title(title)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

## üîç Compare 2D Noise

```python
wavelet_2d = load_image('wavelet_2d.bmp')
perlin_2d = load_image('perlin_2d.bmp')
show_side_by_side(wavelet_2d, perlin_2d, 'Wavelet 2D', 'Perlin 2D')
```

## üîç Compare 3D Noise (Slice and Projected)

```python
wavelet_3d_slice = load_image('wavelet_3d_slice.bmp')
perlin_3d_slice = load_image('perlin_3d_slice.bmp')
wavelet_3d_projected = load_image('wavelet_3d_projected.bmp')

show_side_by_side(wavelet_3d_slice, perlin_3d_slice, 'Wavelet 3D Slice', 'Perlin 3D Slice')
plt.figure(figsize=(5,5))
plt.imshow(wavelet_3d_projected, cmap='gray')
plt.title('Wavelet 3D Projected')
plt.axis('off')
plt.show()
```

## üìä Spectrum Analysis (2D)

```python
spectrum_wavelet_2d = load_image('spectrum_wavelet_2d.bmp')
spectrum_perlin_2d = load_image('spectrum_perlin_2d.bmp')
show_side_by_side(spectrum_wavelet_2d, spectrum_perlin_2d, 'Spectrum Wavelet 2D', 'Spectrum Perlin 2D')
```

## üéö Frequency Bands Comparison

```python
for i in range(3):
    w_band = load_image(f'wavelet_band_{i}.bmp')
    p_band = load_image(f'perlin_band_{i}.bmp')
    show_side_by_side(w_band, p_band, f'Wavelet Band {i}', f'Perlin Band {i}')
```

## üß™ Diagnostic Base Band & Spectra

```python
base = load_image('diagnostic_base_band.bmp')
spectrum = load_image('diagnostic_base_band_spectrum.bmp')
show_side_by_side(base, spectrum, 'Base Band N(x)', 'Base Band Spectrum')
```
