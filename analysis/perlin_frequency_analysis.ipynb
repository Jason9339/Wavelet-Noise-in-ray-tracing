# Perlin Noise Frequency Analysis
This notebook computes the 2D Fourier spectrum of a Perlin noise image to verify that it is not band limited.

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the generated Perlin noise image
img = Image.open('../noise_2d_01.png').convert('L')
data = np.asarray(img, dtype=np.float32) / 255.0
plt.imshow(data, cmap='gray')
plt.title('Perlin noise')
plt.axis('off')
plt.show()
```

```python
# Compute the 2D FFT
fft = np.fft.fftshift(np.fft.fft2(data))
magnitude = np.abs(fft)
log_mag = np.log1p(magnitude)
plt.imshow(log_mag, cmap='inferno')
plt.title('Log magnitude of FFT')
plt.axis('off')
plt.show()
```

The spectrum shows energy spread over many frequencies instead of being contained in a limited band, indicating that the Perlin noise is not band limited.
